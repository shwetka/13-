Наша задачка про лягушку. 
Лягушка упала в канаву 50 (Х) см. Через час она набрала достаточно сил, 
чтобы прыгнуть на 6 ( Y) см, но затем она снова соскользает вниз на 4 (Z) см, и так происходит каждый час. Сколько часов потребуется лягушке, чтобы выбраться из канавы?


```python
print("Введите глубину канавы")
x=int(input())
print("Введите длинну прыжка лягушки")
y=int(input())
print("Введите длинну на которую лягушка соскальзывает")
z=int(input())
if(y-z)>0:
    d=(x-y)/(y-z)+1
    print(d)
elif (x<=y):
    d=1
    print(d)
else:
    print("Произошла Ошибка")
```
